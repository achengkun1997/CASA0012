```python
import pandas as pd
import matplotlib.pylab as plt
```

```python
data = pd.read_csv('hourly 2019.csv')
data
```

```python
plt.figure(figsize=(12, 8))
plt.plot(data['hour'], data['incidents'], marker='.', label='incident')
plt.xticks(range(0, 24))
plt.yticks(range(0, 1800, 200))
plt.legend()
plt.xlabel("Hour")
plt.ylabel('incidents')
plt.title("Hourly change of London Fire Incident in 2019")
for x, y in zip(data['hour'], data['incidents']):
    plt.text(x=x, y=y + 30, s=y)
plt.show()
```
