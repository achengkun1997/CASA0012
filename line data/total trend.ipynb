```python
import pandas as pd
import seaborn as sns
import random
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import matplotlib.patches as mpatches
import numpy as np
import matplotlib.ticker as mtick

%matplotlib inline
sns.set_style('ticks')
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False
```

```python
df = pd.read_csv('monthly change 2016-2020.csv')
df.head(20)
df.columns = ['Month', '2016', '2017', '2018', '2019', '2020']
df = df.reset_index()
df
```

```python
df['Month'].values
```

```python
plt.style.use('ggplot')
f, ax = plt.subplots(figsize=(12, 8))

for col in ['2016', '2017', '2018', '2019', '2020']:
    plt.plot(df['index'], df[col], alpha=0.8, label=col)

ax.set_xticks(range(12))
ax.set_ylim([0, 3500])
ax.set_xticklabels(['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sept', 'Oct', 'Nov','Dec'])

plt.xlabel("Month", fontsize = 14)
plt.ylabel("Incident Number", fontsize = 14)
plt.title('London Fire Incident trend in 2016-2020',fontsize = 15)
plt.legend(ncol=5, loc=2)
plt.show()
```
